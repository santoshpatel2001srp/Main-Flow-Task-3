```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
data = pd.read_csv('D:\\MAIN FLOW Internship\\task-3\\householdtask3.csv')
```

```python
data.head(10)
```

```python
 # Scatter Plot with year against own
plt.scatter(data['year'], data['own'])

# Adding a title to the plot
plt.title('Scatter Plot (year vs own)')

# Setting the x and y labels
plt.xlabel('year')
plt.ylabel('own')

#Showing Plot
plt.show()
```

```python
 # Scatter Plot with tot_hhs against prop_hhs
plt.scatter(data['tot_hhs'], data['prop_hhs'])

# Adding a title to the plot
plt.title('Scatter Plot (tot_hhs vs prop_hhs)')

# Setting the x and y labels
plt.xlabel('tot_hhs')
plt.ylabel('prop_hhs')

#Showing Plot
plt.show()
```

```python
# Line Chart with own_wm against own_prop
plt.plot(data['own_wm'])
plt.plot(data['own_prop'])

# Adding a title to the plot
plt.title('Line Chart (own_wm vs own_prop)')

# Setting the x and y labels
plt.xlabel('own_wm')
plt.ylabel('own_prop')

#Showing Plot
plt.show()
```

```python
# Line Chart with own_wm_prop against prop_hhs
plt.plot(data['own_wm_prop'])
plt.plot(data['prop_hhs'])

# Adding a title to the plot
plt.title('Line Chart (own_wm_prop vs prop_hhs)')

# Setting the x and y labels
plt.xlabel('own_wm_prop')
plt.ylabel('prop_hhs')

#Showing Plot
plt.show()
```

```python
# Bar Chart with year against own
plt.bar(data['year'], data['own'])

# Adding a title to the plot
plt.title('Bar Chart (year vs own)')

# Setting the x and y labels
plt.xlabel('year')
plt.ylabel('own')

#Showing Plot
plt.show()
```

```python
# Bar Chart with age against size
plt.bar(data['age'], data['size'])

# Adding a title to the plot
plt.title('Bar Chart (age vs size)')

# Setting the x and y labels
plt.xlabel('age')
plt.ylabel('size')

#Showing Plot
plt.show()
```
